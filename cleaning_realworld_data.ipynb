```python
# height handling extracts all heights befor " / " and also replaces cm with [space] and
# changes the Dtype of column from object to numeric
df["height_cm"] = df["Measurements"].str.split(" / ").str[0].str.replace(" cm", "") \
   .apply(pd.to_numeric, errors='coerce')
# Weight handling this is different I stucked little bit here like there was only weight values in some rows
# so i am getting NaN in weight cloumn for that problem I took litte help from AI
df["weight_kg"] = df["Measurements"].apply(lambda x: x.split(" / ")[1] if isinstance(x, str) and " / " in x else x) \
    .str.replace(" kg", "").apply(pd.to_numeric, errors='coerce')
```

```python
# Check to make sure all the values contains ' / ' or not
df[~df["Measurements"].str.contains("/", na=False) & df["Measurements"].notna()].head(10)

# remove all cm and kg - removed
```

### Parse out dates from 'Born' and 'Died' columns

```python
df.iloc[91, 4].split(" in ")[0]
```

```python
# Checking for rows which only contains date but no place without NaN value rows
df[~df["Born"].str.contains("in", na=False) & df["Born"].notna()]
```

```python
# Extracting birhtdate using reg exp
date_pattern = r'(\d+ \w+ \d{4}|\d{4})'
df['born_date'] = df['Born'].str.extract(date_pattern)
df['born_year'] = df['Born'].str.extract(r'(\d{4})')

df['born_date'] = pd.to_datetime(df['born_date'], format="mixed", errors='coerce')
df['born_year'] = pd.to_numeric(df['born_year'], errors='coerce')

df['died_date'] = df['Died'].str.extract(date_pattern)
df['died_date'] = pd.to_datetime(df['died_date'], format="mixed", errors='coerce')
```
